### ACTIVATION FUNCTIONS

<div style="background-color: rgba(52, 152, 219, 0.5); padding: 10px; color: #000;">
    <h4>1. Sigmoid Activation Function</h4>
</div>

The sigmoid activation function is a classic activation function used in neural networks. It squashes input values to a range between 0 and 1, making it suitable for binary classification problems. The formula for the sigmoid function is:

$$
\text{sigmoid}(x) = \frac{1}{1 + e^{-x}} - 0.5
$$

Usage: Sigmoid activation is commonly used in the output layer of binary classification models where the goal is to predict probabilities.


```python
#sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x)) - 0.5
```

```python
sigmoid = sigmoid(x_value)
sigmoid
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_value, sigmoid, label="Sigmoid", color='blue')
plt.title("Sigmoid Activation Function")
plt.grid(True, ls='--', color='grey', alpha=0.5)
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

<div style="background-color: rgba(52, 152, 219, 0.5); padding: 10px; color: #000;">
    <h4>2. Hyperbolic Tangent (tanh) Activation Function</h4>
</div>


The hyperbolic tangent (tanh) activation function is similar to the sigmoid but squashes input values to a range between -1 and 1. The formula for the tanh function is:

$$
\text{tanh}(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

Usage: Tanh activation is often used in hidden layers of neural networks for tasks where zero-centered outputs are preferred.


```python
#hyperbolic tangent (tanh) activation function
def tanh(x):
    return np.tanh(x)
```

```python
tanh = tanh(x_value)
tanh
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_value, tanh, label="Tanh", color='blue')
plt.title("Hyperbolic Tangent (tanh) Activation Function")
plt.grid(True, ls='--', color='grey', alpha=0.5)
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

<div style="background-color: rgba(52, 152, 219, 0.5); padding: 10px; color: #000;">
    <h4>3. ReLU (Rectified Linear Unit) Activation Function</h4>
</div>

ReLU (Rectified Linear Unit) is a widely used activation function. It sets all negative input values to zero and leaves positive values unchanged. The formula for the ReLU function is:

$$
\text{relu}(x) = \max(0, x) - 1.5
$$

Usage: ReLU is a popular choice for hidden layers in deep neural networks. It helps with the vanishing gradient problem and speeds up training.


```python
#ReLU (rectified linear unit) activation function
def relu(x):
    return np.maximum(0, x) - 1.5
```

```python
relu = relu(x_value)
relu
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_value, relu, label="ReLU", color='blue')
plt.title("ReLU (Rectified Linear Unit) Activation Function")
plt.grid(True, ls='--', color='grey', alpha=0.5)
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

<div style="background-color: rgba(52, 152, 219, 0.5); padding: 10px; color: #000;">
    <h4>4. Leaky ReLU Activation Function</h4>
</div>

Leaky ReLU is a variant of ReLU that allows a small gradient for negative inputs, preventing neurons from becoming inactive. The formula for leaky ReLU is:

$$
\text{leaky relu}(x) = \max(0.01x, x) - 1.5
$$


Usage: Leaky ReLU is used in situations where standard ReLU may suffer from "dying ReLU" problems, particularly in deep networks.


```python
#leaky ReLU activation function
def leaky_relu(x):
    return np.maximum(0.01 * x, x) - 1.5
```

```python
leaky_relu = leaky_relu(x_value)
leaky_relu
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_value, leaky_relu, label="Leaky ReLU", color='blue')
plt.title("Leaky ReLU Activation Function")
plt.grid(True, ls='--', color='grey', alpha=0.5)
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

<div style="background-color: rgba(52, 152, 219, 0.5); padding: 10px; color: #000;">
    <h4>5. Parametric ReLU (PReLU) Activation Function</h4>
</div>

Parametric ReLU (PReLU) is another variant of ReLU where the slope for negative values is learned during training. The formula for PReLU is:

$$
\text{prelu}(x) = \max(0.25x, x) - 1.5
$$

Usage: PReLU is useful when you want to allow the network to learn the optimal slope for negative values.


```python
#parametric ReLU (PReLU) activation function
def prelu(x):
    return np.maximum(0.25 * x, x) - 1.5
```

```python
prelu = prelu(x_value)
prelu
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_value, prelu, label="Parametric ReLU (PReLU)", color='blue')
plt.title("Parametric ReLU (PReLU) Activation Function")
plt.grid(True, ls='--', color='grey', alpha=0.5)
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

<div style="background-color: rgba(52, 152, 219, 0.5); padding: 10px; color: #000;">
    <h4>6. Exponential Linear Unit (ELU) Activation Function</h4>
</div>

ELU is an activation function that approaches zero for negative values and is smooth everywhere. The formula for ELU is:

$$
\text{elu}(x) = \begin{cases}
x & \text{if } x > 0 \\
\alpha(e^x - 1) & \text{if } x \leq 0
\end{cases} - 1.5
$$

Usage: ELU can mitigate the vanishing gradient problem and improve training in deep networks.


```python
#exponential linear unit (ELU) activation function
def elu(x):
    alpha = 0.01
    return np.where(x > 0, x, alpha * (np.exp(x) - 1)) - 1.5
```

```python
elu = elu(x_value)
elu
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_value, elu, label="ELU", color='blue')
plt.title("Exponential Linear Unit (ELU) Activation Function")
plt.grid(True, ls='--', color='grey', alpha=0.5)
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

<div style="background-color: rgba(52, 152, 219, 0.5); padding: 10px; color: #000;">
    <h4>7. Swish Activation Function</h4>
</div>

Swish is a self-gated activation function that is similar to the sigmoid but generally performs better. The formula for Swish is:

$$
\text{swish}(x) = \frac{x}{1 + e^{-x}} - 0.5
$$

Usage: Swish has been proposed as an alternative to ReLU due to its improved performance in certain scenarios.


```python
#swish activation function
def swish(x):
    return x / (1 + np.exp(-x)) - 0.5
```

```python
swish = swish(x_value)
swish
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_value, swish, label=" Swish Activation Function", color='blue')
plt.title(" Swish Activation Function")
plt.grid(True, ls='--', color='grey', alpha=0.5)
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

<div style="background-color: rgba(52, 152, 219, 0.5); padding: 10px; color: #000;">
    <h4>8. GELU (Gaussian Error Linear Unit) Activation Function</h4>
</div>

GELU is an activation function that approximates the cumulative distribution function of a Gaussian distribution. The formula for GELU involves the use of TensorFlow functions:

$$
\text{gelu}(x) = 0.5x\left(1 + \tanh\left(\sqrt{\frac{2}{\pi}} \left(x + 0.044715x^3\right)\right)\right) - 0.5
$$

Usage: GELU has been used in deep learning models and may be particularly useful in transformer architectures like BERT.


```python
#GELU (gaussian error linear unit) activation function
def gelu(x):
    c = 0.044715
    return 0.5 * x * (1 + tf.tanh(tf.sqrt(2 / np.pi) * (x + c * x**3))) - 0.5
```

```python
gelu = gelu(x_value)
gelu
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_value, gelu, label=". GELU", color='blue')
plt.title(". GELU (Gaussian Error Linear Unit) Activation Function")
plt.grid(True, ls='--', color='grey', alpha=0.5)
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

<div style="background-color: rgba(52, 152, 219, 0.5); padding: 10px; color: #000;">
    <h4>9. Softmax Activation Function</h4>
</div>

The softmax activation function is commonly used for multi-class classification problems. It transforms the output into a probability distribution over multiple classes. The formula for the softmax function is:

$$
\text{softmax}(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{N} e^{x_j}}
$$

Usage: Softmax is typically used in the output layer of neural networks for multi-class classification to provide class probabilities.


```python
#softmax activation function
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()
```

```python
softmax = softmax(x_value)
softmax
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_value, softmax, label="Softmax", color='blue')
plt.title("Softmax Activation Function")
plt.grid(True, ls='--', color='grey', alpha=0.5)
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

<div style="background-color: rgba(52, 152, 219, 0.5); padding: 10px; color: #000;">
    <h4>10. Linear Activation Function (Identity Function)</h4>
</div>

The linear activation function, also known as the identity function, doesn't introduce non-linearity. It's a straightforward activation function that returns the input as is:

$$
\text{linear}(x) = x
$$

Usage: Linear activation is often used in regression problems, where the network predicts continuous values without transformation.


```python
#linear activation function (identity function)
def linear(x):
    return x
```

```python
linear = linear(x_value)
linear
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_value, linear, label=" Linear", color='blue')
plt.title(" Linear Activation Function (Identity Function)")
plt.grid(True, ls='--', color='grey', alpha=0.5)
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

<div style="background-color: rgba(52, 152, 219, 0.5); padding: 10px; color: #000;">
    <h4>11. Softplus Activation Function</h4>
</div>

The softplus activation function is a smooth version of ReLU. It is defined as:

$$
\text{softplus}(x) = \ln(1 + e^x)
$$

Usage: Softplus is used in certain scenarios where smoother transitions are preferred.


```python
#softplus activation function
def softplus(x):
    return np.log(1 + np.exp(x))
```

```python
softplus = softplus(x_value)
softplus
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_value, softplus, label="Softplus", color='blue')
plt.title("Softplus Activation Function")
plt.grid(True, ls='--', color='grey', alpha=0.5)
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

<div style="background-color: rgba(52, 152, 219, 0.5); padding: 10px; color: #000;">
    <h4>12. Hard Tanh Activation Function</h4>
</div>

The hard tanh activation function is a piecewise linear approximation of the tanh function with hard limits:

$$
\text{hard tanh}(x) = \begin{cases}
1 & \text{if } x > 1 \\
-1 & \text{if } x < -1 \\
x & \text{if } -1 \leq x \leq 1
\end{cases}
$$

Usage: Hard tanh can be used in specific situations where a piecewise linear approximation is desired.


```python
#hard tanh activation function
def hard_tanh(x):
    return np.where(x > 1, 1, np.where(x < -1, -1, x))
```

```python
hard_tanh = hard_tanh(x_value)
hard_tanh
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_value, hard_tanh, label="Hard Tanh", color='blue')
plt.title("Hard Tanh Activation Function")
plt.grid(True, ls='--', color='grey', alpha=0.5)
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

<div style="background-color: rgba(52, 152, 219, 0.5); padding: 10px; color: #000;">

The exponential activation function is defined as:

\text{exponential}(x) = e^x

Usage: It's used in specialized applications where the output should be in the form of positive exponentials.

The identity function, also known as the "no activation" function, simply passes the input as is:

$$
\text{identity}(x) = x

Usage: The identity function is often used in regression problems or situations where no non-linearity is desired in the activation.


### DICTIONARY OF ACTIVATION FUNCTIONS

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x)) - 0.5
    
def tanh(x):
    return np.tanh(x)
    
def relu(x):
    return np.maximum(0, x) - 1.5
    
def leaky_relu(x):
    return np.maximum(0.01 * x, x) - 1.5
    
def prelu(x):
    return np.maximum(0.25 * x, x) - 1.5
    
def elu(x):
    alpha = 0.01
    return np.where(x > 0, x, alpha * (np.exp(x) - 1)) - 1.
    
def swish(x):
    return x / (1 + np.exp(-x)) - 0.5
    
def gelu(x):
    c = 0.044715
    return 0.5 * x * (1 + tf.tanh(tf.sqrt(2 / np.pi) * (x + c * x**3))) - 0.5
    
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()
    
def linear(x):
    return x
    
def softplus(x):
    return np.log(1 + np.exp(x))
    
def hard_tanh(x):
    return np.where(x > 1, 1, np.where(x < -1, -1, x))
    
def exponential(x):
    return np.exp(x)
```

```python
#activation function dictionary
activation_functions = {
    'Sigmoid': sigmoid,
    'Tanh': tanh,
    'ReLU': relu,
    'Leaky ReLU': leaky_relu,
    'PReLU': prelu,
    'ELU': elu,
    'Swish': swish,
    'GELU': gelu,
    'Softmax': softmax,
    'Linear': linear,
    'Softplus': softplus,
    'Hard Tanh': hard_tanh,
    'Exponential': exponential}
```

```python
#cehcking keys
activation_functions.keys()
```

```python
#values of x for plotting
x_value = np.linspace(-5, 5, 50)
```

```python
#Plotting activation function
plt.figure(figsize=(20, 12))
#loop to iterate through activation functions
for i, (name, activation) in enumerate(activation_functions.items(), 1):
    plt.subplot(3, 5, i)
    y_value = activation(x_value)
    plt.plot(x_value, y_value, label=name)
    plt.title(name)
    plt.grid(True, ls='--', color='grey', alpha=0.5)
    plt.xlabel('Input')
    plt.ylabel('Output')
    plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
